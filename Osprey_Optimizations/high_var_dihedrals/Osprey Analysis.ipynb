```python
from osprey.config import Config
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
config = Config('config.yaml')
```

```python
df = config.trial_results()
```

```python
df.head()
```

```python
df.sort_values(by='mean_test_score', inplace=True, ascending=False)
```

```python
sns.distplot(df['mean_test_score'])
```

```python
df.head()
```

```python
df['test_err'] = df['test_scores'].apply(lambda x: np.array(x).std())
df['train_err'] = df['train_scores'].apply(lambda x: np.array(x).std())
```

```python
df['rank'] = np.arange(df.shape[0])+1
```

```python
plt.errorbar(df['rank'], df['mean_test_score'], yerr=df['test_err'], fmt='o', alpha=1, ms=5, color='k')

plt.errorbar(df['rank'], df['mean_train_score'], yerr=df['train_err'], fmt='o', alpha=1, ms=5, color='g')
plt.xlim((0,11.5))
plt.ylim((1.8,2.0))
```

```python
params = ['tica__lag_time', 'tica__n_components', 'cluster__n_clusters']
for param in params:
    df[param] = df['parameters'].apply(lambda x: x[param])
```

```python
df.head()
```

```python
plt.errorbar(df['rank'], df['mean_test_score'],yerr=df['test_err'],  color='b', label='test score')
plt.errorbar(df['rank'], df['mean_train_score'], yerr=df['train_err'], color='g', label='train score')
plt.xlim((0,101))
plt.ylim((1.8,2))
plt.ylabel('Model score')
plt.xlabel('Mdel rank')
plt.legend()
```
